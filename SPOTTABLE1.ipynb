1)Given a company name (similar to ones present in companyName), return the close competitors of the company, using any of the fields (or combination of fields) from the dataset. For example, competitors of Swiggy would be Zomato, FoodPanda, DoorDash, etc. Since all of them operate in similar domains and offer similar services

```python
import pandas as pd

# Read the JSON file into a DataFrame
df = pd.read_json('D:\\Data\\domainTags2.json')

# Clean the companyName column
df['companyName'] = df['companyName'].apply(lambda x: x[0])

# Drop rows with null values in the tags column
df = df.dropna(subset=['tags'])

# Prompt the user to enter a company name
company_name = input("Enter a company name to find its top competitors: ")

# Filter the DataFrame to find the row(s) that match the company name
company_row = df.loc[df['companyName'] == company_name]

if len(company_row) == 0:
    print(f"No company found with name '{company_name}'.")
else:
    # Extract the tags associated with the company
    company_tags = company_row.iloc[0]['tags']

    # Find the competitors of the company based on shared tags
    competitors = []
    for _, row in df.iterrows():
        if row['companyName'] == company_name:
            continue
        competitor_tags = row['tags']
        shared_tags = set(company_tags) & set(competitor_tags)
        if len(shared_tags) > 0:
            for tag in company_tags:
                if tag in shared_tags:
                    competitors.append((row['companyName'], tag))
                    break

    # Print the top 5 competitors for each tag
    if len(competitors) == 0:
        print(f"No competitors found for '{company_name}'.")
    else:
        print(f"Top competitors of '{company_name}':")
        tags = set(company_tags)
        for tag in tags:
            tag_competitors = [(c, t) for c, t in competitors if t == tag]
            if len(tag_competitors) > 0:
                sorted_competitors = sorted(tag_competitors, key=lambda x: x[0])
                print(f"- Competitors for tag '{tag}': {[c[0] for c in sorted_competitors[:5]]}")
```
